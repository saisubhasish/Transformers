```python
import gradio as gr


def question_answer(context, question):
    return QA(question=question, context=context)

gr.Interface(fn=question_answer, inputs=["text", "text"], outputs=["text", "text"]).launch()
```

```python
summarization = pipeline("summarization")
```

Abstraction: It generates new text to summerize

Extraction: It extracts the information from text

```python
in_text = "Healthcare is a complex and multifaceted field that involves the prevention, treatment, and management of illnesses and injuries. It encompasses a wide range of specialties, including primary care, specialist care, and emergency medicine. Healthcare professionals include doctors, nurses, pharmacists, and other medical experts who work together to provide high-quality care to patients. In recent years, there has been a growing emphasis on improving access to healthcare, reducing healthcare costs, and incorporating advanced technologies such as telemedicine and artificial intelligence."

print(in_text,"\n\n")
summarization(in_text)[0]["summary_text"]
```
